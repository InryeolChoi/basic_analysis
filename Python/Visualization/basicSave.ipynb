# 그래프 저장법
* 그래프를 저장하는 방법을 생각해보자.
* 생각보다 중요한 작업이 될 수도 있다.

```python
import matplotlib.pyplot as plt

# 예시 데이터
names = ['a', 'b', 'c', 'd', 'e', 'f']
score = [90, 80, 40, 68, 93, 70]

# 그래프 그리기
plt.plot(names, score, 'o', markersize=15)
for n, s in zip(names, score):
    plt.text(n, s, str(s), fontsize=12, ha='right')

plt.xlabel('Names')
plt.ylabel('Scores')
plt.title('Scatter Plot with Names on X-axis')
plt.show()
```
